###Python OOP: Core Concepts

Object-Oriented Programming (OOP) is a programming paradigm based on classes and objects. It helps organize code into reusable, modular, and maintainable structures.

**Classes and Objects**



1.   A class is a blueprint for creating objects.

2.   An object is an instance of a class.



```python
class Dog:
    # Constructor (initializer)
    def __init__(self, name, breed):
        self.name = name      # attribute
        self.breed = breed    # attribute
        sum = 2 + 2
        print(sum)

    # Method
    def bark(self):
        print(f"{self.name} says Woof!")

# Create an object
my_dog = Dog("Buddy", "Golden Retriever")
# my_dog.bark()  # Output: Buddy says Woof!
```

```python
class ATM():
    def __init__(self, name, password, balance=0):
        self.balance = balance
        self.name = name
        self.password = password

    def check_balance(self, password):
        if password == self.password:
            print(f"Your balance is {self.balance}")
        else:
            print("Wrong password")

    def withdraw(self, amount, password):
        if password != self.password:
            print("Wrong password")
            return
        if amount > self.balance:
            print("Not enough money")
        else:
            self.balance -= amount
            print(f"You have withdrawn {amount}. Your new balance is {self.balance}")

# Create account
my_account = ATM("Zain", 1234, 1000)
# my_account.check_balance(1234)

# # Now call methods with password
# my_account.check_balance(1234)          # Correct password
my_account.withdraw(200, 1234)         # Correct password
# my_account.check_balance(9999)         # Wrong password
```

```python
# Example 2

class TextAnalyzer:
    def __init__(self, text):
        self.text = text  # string input

    def word_count(self):
        return len(self.text.split())

    def char_count(self):
        return len(self.text)

    def to_upper_manual(self):  # without using .upper()
        result = ""
        for char in self.text:
            if 'a' <= char <= 'z':
                result += chr(ord(char) - 32)
            else:
                result += char
        return result

    def to_lower_manual(self):  # without using .lower()
        result = ""
        for char in self.text:
            if 'A' <= char <= 'Z':
                result += chr(ord(char) + 32)
            else:
                result += char
        return result
# Usage
# sentence = TextAnalyzer("hello world from zain ALI")
# print(sentence.word_count())      # 4
# print(sentence.char_count())      # 21
# print(sentence.to_upper_manual()) # HELLO WORLD FROM ZAIN
```

```python
sentence = TextAnalyzer("hello world from zain ALI")
print(sentence.word_count())
print(sentence.char_count())
print(sentence.to_upper_manual())
print(sentence.to_lower_manual())
```

```python
# Example 3

class NumberList:
    def __init__(self, numbers):
        self.numbers = numbers  # list input

    def total(self):
        return sum(self.numbers)

    def average(self):
        if self.numbers:
          average = sum(self.numbers) / len(self.numbers)
          return average
        else:
          return 0

    def add_number(self, num):
        self.numbers.append(num)
        return self.numbers

    def get_even_numbers(self):
        return [n for n in self.numbers if n % 2 == 0]

# Usage
nums = NumberList([2, 4, 5, 7, 8])
print(nums.total())           # 26
print(nums.average())         # 5.2
print(nums.add_number(10))
print(nums.get_even_numbers()) # [2, 4, 8, 10]
```

```python
# Example 4

class StudentGrades:
    def __init__(self, grades_dict):
        self.grades = grades_dict  # dict: {'name': score}

    def add_student(self, name, score):
        self.grades[name] = score

    def get_top_student(self):
        if not self.grades:
            return None
        else:
            top_student = max(self.grades, key=self.grades.get)
            return top_student

    def average_score(self):
        if not self.grades:
            return 0
        return sum(self.grades.values()) / len(self.grades)

    def get_all_names(self):
        return list(self.grades.keys())

# Usage
grades = StudentGrades({"Zain": 92, "Ali": 88, "Sara": 95})
grades.add_student("Layla", 89)
print(grades.get_top_student())  # Sara
print(grades.average_score())    # 91.0
print(grades.get_all_names())    # ['Zain', 'Ali', 'Sara', 'Layla']
```

**Encapsulation**

Encapsulation means bundling data (attributes) and the methods that operate on that data inside a single unit (class), and controlling access to that data to prevent accidental misuse.

Hide internal details, expose only what’s necessary.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age # Private property

p1 = Person("Emil", 25)
print(p1.name)
print(p1.__age) # This will cause an error
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name        # public (okay for name)
        self.__age = None       # private
        self.set_age(age)       # use setter for validation

    def set_age(self, age):
        if age < 0:
            raise ValueError("Age cannot be negative!")
        self.__age = age

    def get_age(self):
        return self.__age

# Usage
s = Student("Zain", 22)
print(s.get_age())  # 22

s.set_age(25)       # OK
# s.__age = -10     # ❌ Won't affect real __age (creates new public __age!)
s.get_age()
```

```python
class BankAccount:
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder  # public (okay to read)
        self.__balance = 0                    # private — protected!
        self.__transaction_history = []       # private — no outside access
        self.deposit(initial_balance)         # use method for validation

    # 👇 Encapsulated balance access using @property (Pythonic way)
    @property
    def balance(self):
        """Secure read-only access to balance."""
        return self.__balance



    def deposit(self, amount):
        """Add money to account with validation."""
        if amount <= 0:
            raise ValueError("Deposit amount must be positive!")
        self.__balance += amount
        self.__transaction_history.append(f"Deposited: +{amount}")
        print(f"✅ Deposited ${amount}. New balance: ${self.__balance}")

    def withdraw(self, amount):
        """Withdraw money if sufficient funds."""
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive!")
        if amount > self.__balance:
            raise ValueError("Insufficient funds!")
        self.__balance -= amount
        self.__transaction_history.append(f"Withdrew: -{amount}")
        print(f"✅ Withdrew ${amount}. New balance: ${self.__balance}")

    def get_transaction_history(self):
        """Return a copy of transaction history (not the original list!)."""
        return self.__transaction_history.copy()

    def __str__(self):
        return f"Account of {self.account_holder} | Balance: ${self.__balance}"
```

```python
# Create account
acc = BankAccount("Zain", 100)

# Safe access
print(acc.account_holder)  # Zain (public)
print(acc.balance)         # 100 (via @property — read-only)

# Valid operations
acc.deposit(50)    # ✅ Deposited $50. New balance: $150
acc.withdraw(30)   # ✅ Withdrew $30. New balance: $120

# ❌ These will FAIL or not work as expected:

# acc.balance = 9999        # ❌ AttributeError (no setter defined → read-only)
# acc.__balance = -500      # ❌ Creates fake public __balance — real balance unchanged!
# print(acc.__transaction_history)  # ❌ AttributeError (private)

# Safe history access
print("\nTransaction History:")
for log in acc.get_transaction_history():
    print(" -", log)
```
